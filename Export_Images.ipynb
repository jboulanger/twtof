# Export FIB image and peak to TIF

```python
import twtof
import tifffile
import numpy as np

filename = "./data/20230722_magnetotactic_HO027_1_Au_pos_spot2_run1_30kV_50pA.h5"

fib, peak = twtof.imread(filename)

tifffile.imwrite(
    "mass.tif",
    np.swapaxes(peak, 0, 1),
    imagej=True,
    resolution=(1.0 / twtof.get_scale(peak,"X"), 1.0 / twtof.get_scale(peak,"Y")),
    metadata={"axes": "ZCYX"},
)

tifffile.imwrite(
    "fib.tif", fib.astype(np.uint16), imagej=True, metadata={"axes": "ZYX"}
)
```
